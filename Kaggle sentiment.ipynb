```python
# Keeping only the neccessary columns
data1 = data[['title','body','rating','product_handle','product_id']]
```

```python
data1.head(5)
```

```python
# Splitting the dataset into train and test set
train, test = train_test_split(data1,test_size = 0.1)
```

```python
# Removing neutral sentiments
train = train[train.rating != "Neutral"]
```

```python
train_pos = train[ train['rating'] == 'Positive']
train_pos = train_pos['body']
train_neg = train[ train['rating'] == 'Negative']
train_neg = train_neg['body']

def wordcloud_draw(data, color = 'black'):
    words = ' '.join(data)
    cleaned_word = " ".join([word for word in words.split()
                            if 'http' not in word
                                and not word.startswith('@')
                                and not word.startswith('#')
                                and word != 'RT'
                            ])
    wordcloud = WordCloud(stopwords=STOPWORDS,
                      background_color=color,
                      width=2500,
                      height=2000
                     ).generate(cleaned_word)
    plt.figure(1,figsize=(13, 13))
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.show()
    
print("Positive words")
wordcloud_draw(train_pos,'white')
print("Negative words")
wordcloud_draw(train_neg)
```

```python
# assign reviews with score > 3 as positive sentiment
# score < 3 negative sentiment
# remove score = 3
data1 = data1[data1['rating'] != 3]
data1['sentiment'] = data1['rating'].apply(lambda rating : +1 if rating > 3 else -1)
```

```python
# split df - positive and negative sentiment:
positive = data1[data1['sentiment'] == 1]
negative = data1[data1['sentiment'] == -1]
```

```python
data1['sentiment1'] = data1['sentiment'].replace({-1 : 'negative'})
data1['sentiment1'] = data1['sentiment'].replace({1 : 'positive'})
```
